```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 데이터 불러오기
df_mid = pd.read_csv('data/서울시 중학교 기본정보.csv', encoding='cp949')
df_high = pd.read_csv('data/서울시 고등학교 기본정보.csv', encoding='cp949')
df_open = pd.read_csv('data/서울특별시교육청_민간개방 학교시설물현황_20250910.csv', encoding='cp949')
```

```python
# 중복값 제거하기
df_mid.drop_duplicates(inplace=True)
df_high.drop_duplicates(inplace=True)
```

```python
df_mid.head()
```

```python
df_high.head()
```

```python
# 중학교 개수 확인
num_mid = df_mid['학교명'].nunique()
print(f'서울시 중학교 개수: {num_mid}')

# 고등학교 개수 확인
num_high = df_high['학교명'].nunique()
print(f'서울시 고등학교 개수: {num_high}')
```

```python
df_open.head()
```

```python
# 중학교 중 공립/사립 개수 확인
mid_public = df_mid[df_mid['설립구분'] == '공립']['학교명'].nunique()
mid_private = df_mid[df_mid['설립구분'] == '사립']['학교명'].nunique()
print(f'서울시 중학교 공립 개수: {mid_public}')
print(f'서울시 중학교 사립 개수: {mid_private}')
# 고등학교 중 공립/사립 개수 확인
high_public = df_high[df_high['설립구분'] == '공립']['학교명'].nunique()
high_private = df_high[df_high['설립구분'] == '사립']['학교명'].nunique()
print(f'서울시 고등학교 공립 개수: {high_public}')
print(f'서울시 고등학교 사립 개수: {high_private}')
```

```python
# 데이터 합치기
df_combined = pd.concat([df_mid, df_high], ignore_index=True)
df_combined.head()
```

```python
# 민간개방 데이터의 '금호여자중학교'를 '금호중학교'로 변경
df_open['학교명'] = df_open['학교명'].replace('금호여자중학교', '금호중학교')

# 민간개방 데이터와 합쳐진 데이터에서 학교명 기준으로 일치하는 학교 개수 확인
open_schools = df_open['학교명'].unique()
combined_schools = df_combined['학교명'].unique()
matching_schools = set(open_schools) & set(combined_schools)
print(f'민간개방 데이터와 합쳐진 데이터에서 일치하는 학교 개수: {len(matching_schools)}')
```

```python
# 민간개방 데이터와 합쳐진 데이터를 합치기
df_final = pd.merge(df_combined, df_open, on='학교명', how='inner')
df_final.head()
```

```python
# 중학교 개수 세기
num_mid_final = df_final[df_final['학교명'].isin(df_mid['학교명'])]['학교명'].nunique()
print(f'최종 데이터에서 중학교 개수: {num_mid_final}')
# 중학교 중에서도 사립과 공립 개수 세기
num_mid_public_final = df_final[(df_final['학교명'].isin(df_mid[df_mid['설립구분'] == '공립']['학교명']))]['학교명'].nunique()
num_mid_private_final = df_final[(df_final['학교명'].isin(df_mid[df_mid['설립구분'] == '사립']['학교명']))]['학교명'].nunique()
print(f'최종 데이터에서 중학교 공립 개수: {num_mid_public_final}')
print(f'최종 데이터에서 중학교 사립 개수: {num_mid_private_final}')
# 고등학교 개수 세기
num_high_final = df_final[df_final['학교명'].isin(df_high['학교명'])]['학교명'].nunique()
print(f'최종 데이터에서 고등학교 개수: {num_high_final}')
# 고등학교 중에서도 사립과 공립 개수 세기
num_high_public_final = df_final[(df_final['학교명'].isin(df_high[df_high['설립구분'] == '공립']['학교명']))]['학교명'].nunique()
num_high_private_final = df_final[(df_final['학교명'].isin(df_high[df_high['설립구분'] == '사립']['학교명']))]['학교명'].nunique()
print(f'최종 데이터에서 고등학교 공립 개수: {num_high_public_final}')
print(f'최종 데이터에서 고등학교 사립 개수: {num_high_private_final}')
```

```python
# 운동장 개방여부=유인 학교 개수 세기
num_open = df_final[df_final['운동장 개방여부'] == '유']['학교명'].nunique()
```

```python
# 출력
print(f'최종 데이터에서 운동장 개방여부가 유인 학교 개수: {num_open}')
```

```python
# 이 중 중학교 개수 세기
num_open_mid = df_final[(df_final['운동장 개방여부'] == '유') & (df_final['학교명'].isin(df_mid['학교명']))]['학교명'].nunique()
# 이 중 고등학교 개수 세기
num_open_high = df_final[(df_final['운동장 개방여부'] == '유') & (df_final['학교명'].isin(df_high['학교명']))]['학교명'].nunique()
# 출력
print(f'최종 데이터에서 운동장 개방여부가 유인 중학교 개수: {num_open_mid}')
print(f'최종 데이터에서 운동장 개방여부가 유인 고등학교 개수: {num_open_high}')
```
